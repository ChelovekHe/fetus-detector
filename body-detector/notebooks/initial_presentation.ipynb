Automated localisation of the fetal thorax in MRI
=================================================

* the orientation is found by registering the motion corrected brain to an atlas
  * Ison et al. used tissue segmentation to find the orientation
  * alternative could be to make tests in a local basis pointing toward the heart
* RF forest are successively applied: classification, regression, classification, regression
  * this is a simplified approach of the Hough Forests
  * the second step of classification/regression uses autocontext: features extracted from the output of the previous classifiers

Reveal.js parameters: ?transition=none&width=1200&slideNumber=true

```python
def scores_to_dict(scores,i):
    res = {}
    for s in scores:
        res[str(int(s[0]))] = s[i]
    return res

def random_pick(choices, probs, n):
    '''
    http://stackoverflow.com/questions/8973350/draw-random-element-in-numpy
    >>> a = ['Hit', 'Out']
    >>> b = [.3, .7]
    >>> random_pick(a,b)
    '''
    cutoffs = np.cumsum(probs)
    idx = cutoffs.searchsorted(np.random.uniform(0, cutoffs[-1], size=n))
    return choices[idx]
```

```python
def explore_dataset(myfolder,verbose=False,scores=None):
    all_files = glob( myfolder + "/*_aligned.nii.gz")

    R = []
    theta = []
    ga = []
    X = []
    Y = []
    for f in all_files:
        patient_id = os.path.basename(f)[:-len("_aligned.nii.gz")]
        if scores is not None:
            ga.append( scores[patient_id] )
        else:
            ga.append( all_ga[patient_id] )
        labels = irtk.imread(f)

        brain_center = labels.WorldToImage([0,0,0])[::-1].astype('float32')

        heart_center =  np.array(nd.center_of_mass( (labels == 5).view(np.ndarray) ),
                                 dtype='float32')

        v = heart_center - brain_center
        R.append(np.linalg.norm(v))

        # A . B = |A||B|cos(angle)
        angle = np.dot(v,np.array([-1,0,0],dtype='float32'))/np.linalg.norm(v)
        angle = np.arccos(angle) / np.pi * 180.0
        theta.append(angle)

        X.append(v[2])
        Y.append(v[1])

        if verbose:
            print patient_id, R[-1], theta[-1]

    ## R & theta
    fig, ax = plt.subplots(1,2,figsize=(18,8),dpi=300)

    sc = ax[0].scatter(theta, R, c=np.array(ga), cmap=plt.cm.jet, s=50)

    #cbar = plt.colorbar(sc)
    #cbar.set_label('Gestational age', rotation=270)

    ax[0].grid(True)

    ax[0].set_xlabel('theta (degrees)')
    ax[0].set_ylabel('R (px)')
    ax[0].set_title('R-theta variability across dataset')

    ax[0].set_xlim([0,80])
    ax[0].set_ylim([0,140])
    #plt.axis('equal')

    #fig.tight_layout()
    #plt.savefig("data_R_theta_CRL.png")
    #plt.close(fig)

    ## X & Y    
    sc = ax[1].scatter(X, Y, c=np.array(ga), cmap=plt.cm.jet, s=50)

    cbar = plt.colorbar(sc)
    if scores is not None:
        cbar.set_label('Score', rotation=270)
    else:
        cbar.set_label('Gestational age', rotation=270)

    ax[1].grid(True)

    ax[1].set_xlabel('x - left to right (px)')
    ax[1].set_ylabel('y - back to front (px)')
    ax[1].set_title('Heart location in transverse plane (brain orientation)')

    ax[1].set_xlim([-75,75])
    ax[1].set_ylim([-75,75])
    ax[1].set_aspect('equal')

    #fig.tight_layout()
    #plt.savefig("data_heart_XY_CRL.png")
    #plt.close(fig)
```

```python
def dummy_predictor(myfolder):
    all_files = glob( myfolder + "/*_aligned.nii.gz")

    offset = []
    for f in all_files:
        patient_id = os.path.basename(f)[:-len("_aligned.nii.gz")]
        labels = irtk.imread(f)

        brain_center = labels.WorldToImage([0,0,0])[::-1].astype('float32')

        heart_center =  np.array(nd.center_of_mass( (labels == 5).view(np.ndarray) ),
                                 dtype='float32')
        
        v = heart_center - brain_center

        offset.append(v)
        
    offset = np.mean(offset,axis=0)
    
    scores = {}
    for f in all_files:
        patient_id = os.path.basename(f)[:-len("_aligned.nii.gz")]
        labels = irtk.imread(f)

        brain_center = labels.WorldToImage([0,0,0])[::-1].astype('float32')

        heart_center =  np.array(nd.center_of_mass( (labels == 5).view(np.ndarray) ),
                                 dtype='float32')
        
        scores[patient_id] = np.linalg.norm( brain_center + offset - heart_center )

    return scores
```

```python
def score_mean_shift_one_patient( detection_folder,
                          ground_truth_folder,
                          patient_id,
                      bandwidth=20,
                      n_points=1000,
                      verbose=True):
    
    labels = irtk.imread( ground_truth_folder + "/" + patient_id + "_aligned.nii.gz")
    center =  np.array(nd.center_of_mass( np.logical_or(np.logical_or(labels == 3,
                                                                           labels == 4),
                                                             labels == 5).view(np.ndarray)),dtype='float32')
        
    img = irtk.imread( detection_folder+"/"+patient_id+"_img.nii.gz" )
    votes_heart = irtk.imread( detection_folder+"/"+patient_id+"_votes_blurred_heart.nii.gz")
    votes_lungs = irtk.imread( detection_folder+"/"+patient_id+"_votes_blurred_lungs.nii.gz")
        
    points_lungs = np.argwhere(votes_lungs)
    probas_lungs = votes_lungs[points_lungs[:,0],
                         points_lungs[:,1],
                         points_lungs[:,2]]
    #points_lungs = random_pick(points_lungs,probas,n_points)

    points_heart = np.argwhere(votes_heart)
    probas_heart = votes_heart[points_heart[:,0],
                         points_heart[:,1],
                         points_heart[:,2]]
    #points_heart = random_pick(points_heart,probas,n_points)

    #points = np.concatenate((points_lungs,points_heart))
    
    points = np.concatenate((points_lungs,points_heart))
    probas = np.concatenate((probas_lungs,probas_heart))
    points = random_pick(points,probas,n_points)

    ms = MeanShift(bandwidth=bandwidth)
    ms.fit(points)

    best_sum = 0
    best = -1
    for i,c in enumerate(ms.cluster_centers_):
        s = np.sum(probas[ms.labels_==i])
        if s > best_sum:
            best_sum = s
            best = i
                
    detection = ms.cluster_centers_[best]
        
    s = np.linalg.norm( center - detection )

    inside = labels[detection[0],
                            detection[1],
                            detection[2]] == 5
    
    if verbose:
        print patient_id, s, inside
    else:
        if s > 50:
            print patient_id, s, inside
        
    return [int(patient_id), s, inside]

def score_mean_shift( detection_folder,
                      ground_truth_folder,
                      bandwidth=20,
                      verbose=True,
                      n_jobs=10,
                      n_points=1000):
        
    all_files = glob( detection_folder+"/*_img.nii.gz" )
    
    all_patients = []
    for f in all_files:
        all_patients.append( os.path.basename(f)[:-len("_img.nii.gz")] )
        
    scores = Parallel(n_jobs=n_jobs)(delayed(score_mean_shift_one_patient)( detection_folder,
                                                                            ground_truth_folder,
                                                                            patient_id,
                                                                            bandwidth=bandwidth,
                                                                            verbose=verbose,
                                                                            n_points=n_points)
                                                                    for patient_id in all_patients )
    if verbose:
        print "total:", len(scores), "files"
        
    return np.array(scores)
```

```python
def plot_scores(scores,ground_truth_folder=None):
    
    if ground_truth_folder is not None:
        dummy_scores = dummy_predictor(ground_truth_folder)
        
    x = []
    y = []
    c = []
    dummy = []
    for s in scores:
        x.append( all_ga[str(int(s[0]))] )
        y.append( s[1] )
        if ground_truth_folder is not None:
            dummy.append(dummy_scores[str(int(s[0]))])
        if s[2] == 1:
            c.append('g')
        else:
            c.append('r')
            
    x = np.array(x)
    y = np.array(y)
    if ground_truth_folder is not None:
        dummy = np.array(dummy)

    fig, ax = plt.subplots(figsize=(6,5))
    ax.scatter( x, y, c=c, alpha=0.5)
    if ground_truth_folder is not None:
        #ax.plot( np.sort(x), dummy[np.argsort(x)], c='b', alpha=0.5, linestyle='-')
        ax.axhline(y=np.mean(dummy), c='b', alpha=0.5, linestyle='--',label='dummy y = '+str(np.mean(dummy)))

    plt.xlim(20,40)
    plt.ylim(0.0,200.0)

    ax.grid(True)
    
    legend = ax.legend(loc='upper right')

    plt.xlabel('Gestational age (weeks)')
    plt.ylabel('Error distance (mm)')
    plt.title('Detection of the center of the heart')

    #fig.tight_layout()
    
    return
```

## Center of each lung

```python
def score_detection_lungs( detection_folder,
                      ground_truth_folder,
                      verbose=True):
    all_files = glob( detection_folder+"/*_votes_blurred_heart.nii.gz" )
    all_files = sorted(all_files)
    scores = []
    for f in all_files:
        patient_id = os.path.basename(f).split('_')[0]
        labels = irtk.imread( ground_truth_folder + "/" + patient_id + "_aligned.nii.gz")

        center =  np.array(nd.center_of_mass( (labels == 5).view(np.ndarray) ),dtype='float32')
    
        votes_heart = irtk.imread(f)

        detected_heart = np.unravel_index(np.argmax(votes_heart), labels.shape )
        
        m = np.zeros(votes_heart.shape, dtype='uint8')
        m[detected_heart[0],
          detected_heart[1],
          detected_heart[2]] = 1

        edt = nd.distance_transform_edt(np.logical_not(m))
        votes_heart[edt > 15] = 0
          
        offsets_lungs = irtk.imread( detection_folder+'/'+patient_id+'_offsets_lungs.nii.gz',dtype='int32')

        votes_lungs = crossed_hough_votes(offsets_lungs,votes_heart)
        
        points_lungs = np.argwhere(votes_lungs)
        probas_lungs = votes_lungs[points_lungs[:,0],
                             points_lungs[:,1],
                             points_lungs[:,2]]
        points_lungs = random_pick(points_lungs,probas_lungs,2000)


        kmeans = KMeans(n_clusters=2)
        kmeans.fit(points_lungs)
        
        center_lungs =  np.array(nd.center_of_mass( np.ones(labels.shape),
                                                   labels.view(np.ndarray), index=[3,4] ),dtype='float32')
            
        detection1 = kmeans.cluster_centers_[0]
        detection2 = kmeans.cluster_centers_[1]
        
        s1 = np.argmin( np.linalg.norm( center_lungs - detection1, axis=1 ) )
        s2 = np.argmin( np.linalg.norm( center_lungs - detection2, axis=1 ) )

        inside1 = labels[detection1[0],
                        detection1[1],
                        detection1[2]] == 3+s1
        inside2 = labels[detection2[0],
                        detection2[1],
                        detection2[2]] == 3+s2
        
        s1 = np.linalg.norm( center_lungs - detection1, axis=1 )[s1]
        s2 = np.linalg.norm( center_lungs - detection2, axis=1 )[s2]
    
        scores.append([int(patient_id), s1, s2, inside1, inside2])
        if verbose:
            print patient_id, s1, s2, inside1, inside2
        
    if verbose:
        print "total:", len(scores), "files"
    return np.array(scores)

def plot_scores_lungs(scores,ground_truth_folder=None):
    
    if ground_truth_folder is not None:
        dummy_scores = dummy_predictor(ground_truth_folder)
        
    x = []
    y = []
    c = []
    dummy = []
    for s in scores:
        x.append( all_ga[str(int(s[0]))] )
        x.append( all_ga[str(int(s[0]))] )
        y.append( s[1] )
        y.append( s[2] )
        if ground_truth_folder is not None:
            dummy.append(dummy_scores[str(int(s[0]))])
        if s[3] == 1:
            c.append('g')
        else:
            c.append('r')
        if s[4] == 1:
            c.append('g')
        else:
            c.append('r')
            
    x = np.array(x)
    y = np.array(y)
    c = np.array(c)
    if ground_truth_folder is not None:
        dummy = np.array(dummy)

    fig, ax = plt.subplots(figsize=(6,5))
    # ax.scatter( x, y, c=c, alpha=0.5)
    scat1 = ax.scatter( x[c=='g'], y[c=='g'], c='g', alpha=0.5, label='inside')
    scat2 = ax.scatter( x[c=='r'], y[c=='r'], c='r', alpha=0.5, label='outside')
    
    if ground_truth_folder is not None:
        #ax.plot( np.sort(x), dummy[np.argsort(x)], c='b', alpha=0.5, linestyle='-')
        ax.axhline(y=np.mean(dummy), c='b', alpha=0.5, linestyle='--',label='dummy y = '+str(np.mean(dummy)))

    plt.xlim(20,40)
    plt.ylim(0.0,200.0)

    ax.grid(True)
    
    #if ground_truth_folder is not None:
    legend = ax.legend(loc='upper right')

    plt.xlabel('Gestational age (weeks)')
    plt.ylabel('Error distance (mm)')
    plt.title('Detection of the center of the lung')

    #fig.tight_layout()
    
    return
```

```python
class SKImage(skimage.io.Image):
    """
    http://nbviewer.ipython.org/urls/gist.github.com/stefanv/6416926/raw/96978d6ad6991c26ccce995327b1adb6aceeeb6e/side_by_side.ipynb
    """
    def __init__(self, *args, **kwargs):
        self._width = kwargs.pop('width', '100%')
        skimage.io.Image.__init__(self, *args, **kwargs)
    def _repr_html_(self):
        out = '<img width="%s"' % self._width
        out += ' src="data:image/png;base64,' + \
               base64.b64encode(np.compat.asbytes(self._repr_png_())) + '"/>'
        return out

class MyTable(object):
    def __init__(self, folder,width=100):
        self.folder = folder
        self.width = width

    def _repr_html_(self):
        all_files = glob( self.folder+"/*" )
        s = "<div style='float: left;'>"
        s+= "<table style='width:"+str(self.width)+"%'>"
        for i in range(0,len(all_files)/2,2):
            f1 = 'file://'+os.path.abspath(all_files[2*i])
            if (len(all_files) % 2) == 1:
                f2 = os.path.abspath(all_files[2*i+1])
            else:
                f2 = None
            s +=  "<tr><td>"+SKImage(skimage.io.imread(f1))._repr_html_()+"</td>"
            if f2 is not None:
                s +=  "<td>"+SKImage(skimage.io.imread(f2))._repr_html_()+"</td>"
            s += "</tr>"
            
        s += "</table></div>"

        return s
    
class MyTableClustered(object):
    def __init__(self, folder,width=100):
        self.folder = folder
        self.width = width

    def _repr_html_(self):
        all_files = glob( self.folder+"/*" )
        all_patients = []
        for f in all_files:
            patient_id = os.path.basename(f).split('_')[0]
            if patient_id not in all_patients:
                all_patients.append(patient_id)
        s = "<div style='float: left;'>"
        s+= "<table style='width:"+str(self.width)+"%'>"
        for patient_id in all_patients:
            all_files = glob( self.folder+"/"+patient_id+"_*" )
            s +=  "<tr><td>"+patient_id+"</td>"
            for f in all_files:
                s +=  "<td>"+SKImage(skimage.io.imread(f))._repr_html_()+"</td>"
            s += "</tr>"
            
        s += "</table></div>"
        
        print "Number of patients:", len(all_patients)

        return s
```
